```python
# ==== LIMPIEZA SIMPLE Y PRECISA CON IMPUTACIÓN ====
import pandas as pd
import numpy as np
import re

# 1) Cargar archivo
datos_original = pd.read_excel(FILE_PATH)

# --- Métricas "ANTES" ---
forma_antes = datos_original.shape
nulos_antes = datos_original.isna().sum()
duplicados_antes = datos_original.duplicated().sum()

# 2) Copia de trabajo
datos = datos_original.copy()

# 3) Eliminar columnas con más de 90% de valores nulos
porcentaje_nulos = datos.isna().mean()
columnas_conservar = [c for c in datos.columns if porcentaje_nulos[c] <= 0.90]
columnas_eliminadas = [c for c in datos.columns if c not in columnas_conservar]
datos = datos[columnas_conservar]

# 4) Eliminar filas duplicadas
duplicados_encontrados = datos.duplicated().sum()
datos = datos.drop_duplicates()

# 5) Limpiar cadenas de texto (espacios, guiones, valores vacíos)
for col in datos.columns:
    if datos[col].dtype == object:
        datos[col] = (datos[col].astype(str)
                      .str.strip()
                      .replace({"": np.nan, "-": np.nan, "nan": np.nan,
                                "None": np.nan, "NaN": np.nan}))

# 6) Función para convertir a número de manera segura
def convertir_a_numero(valor):
    if pd.isna(valor):
        return np.nan
    if isinstance(valor, (int, float)):
        return float(valor)
    s = str(valor).strip()
    s = s.replace("S/.", "").replace("S/", "").replace(",", "").replace("\xa0","")
    s = re.sub(r"\s+", "", s)
    try:
        return float(s)
    except:
        return np.nan

# Columnas típicas a convertir (si existen en tu archivo)
columnas_numericas = [
    "porcentaje_propiedad","area_terreno_m2","area_comun_terreno_m2",
    "area_construida_m2","area_comun_construida_m2","area_total_construida_m2",
    "pisos","anio_construccion","mayor_anio_construccion",
    "valor_terreno_2025","valor_construccion_depreciado_2025",
    "valor_obras_complementarias_depreciado_2025","autovaluo_2025"
]

for col in columnas_numericas:
    if col in datos.columns:
        datos[col] = datos[col].apply(convertir_a_numero)

# 7) Reglas mínimas (muy básicas)
for col in ["area_terreno_m2","area_total_construida_m2","pisos",
            "valor_terreno_2025","valor_construccion_depreciado_2025",
            "valor_obras_complementarias_depreciado_2025","autovaluo_2025"]:
    if col in datos.columns:
        datos.loc[datos[col] < 0, col] = np.nan

# 8) Eliminar filas con nulos en columnas críticas
columnas_criticas = [c for c in ["autovaluo_2025","area_terreno_m2"] if c in datos.columns]
filas_antes = datos.shape[0]
if columnas_criticas:
    datos = datos.dropna(subset=columnas_criticas)
filas_eliminadas = filas_antes - datos.shape[0]

# 9) IMPUTAR VALORES FALTANTES (rellenar NaN)
num_cols = datos.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in datos.columns if c not in num_cols]

# Numéricas → mediana
for c in num_cols:
    mediana = datos[c].median()
    datos[c] = datos[c].fillna(mediana)

# Categóricas → "Desconocido"
for c in cat_cols:
    datos[c] = datos[c].fillna("Desconocido")

# --- Métricas "DESPUÉS" ---
forma_despues = datos.shape
nulos_despues = datos.isna().sum()

# 10) Guardar dataset limpio
ARCHIVO_SALIDA = "predios_limpio_simple.csv"
datos.to_csv(ARCHIVO_SALIDA, index=False, encoding="utf-8")

# Resumen de cambios
print("=== RESUMEN DE LIMPIEZA SIMPLE CON IMPUTACIÓN ===")
print(f"Forma ANTES : {forma_antes}")
print(f"Forma DESPUÉS: {forma_despues}")
print(f"Columnas eliminadas (>90% nulos): {len(columnas_eliminadas)} -> {columnas_eliminadas[:10]}")
print(f"Duplicados eliminados: {duplicados_encontrados}")
print(f"Filas eliminadas por nulos críticos {columnas_criticas}: {filas_eliminadas}")
print("Valores nulos restantes (DESPUÉS):")
print(nulos_despues.sum())
print(f"✔ Archivo limpio guardado en: {ARCHIVO_SALIDA}")
```

2. Problema: para entrenar un modelo de clasificación, necesitamos una clase categórica

Los algoritmos de clasificación (como J48 / Árbol de decisión) requieren una variable objetivo (y) que no sea continua, sino categórica (ejemplo: sí/no, alto/bajo, 0/1).

Por eso convertimos el autovalúo numérico en una etiqueta binaria.


🔹 3. ¿Cómo hacemos la conversión?

Calculamos el percentil 75 (Q3) de todos los valores de autovalúo.

Ese número (umbral_q3) significa que el 75% de los predios tienen un autovalúo menor o igual a ese valor, y el 25% más caro está por encima.
 1. Propósito académico / técnico

En tu proyecto, crear la clase target_autovaluo te da:

Un problema de clasificación binaria claro (0/1) para aplicar algoritmos como J48 (árbol de decisión).

La posibilidad de entrenar, evaluar y comparar modelos de machine learning con métricas objetivas (accuracy, matriz de confusión, etc.).

Una base para practicar el preprocesamiento, PCA, y evaluación de modelos en scikit-learn.

Es decir: usar el autovalúo como criterio de clasificación es un ejemplo pedagógico para mostrar todo el flujo de ciencia de datos.

🔹 2. Propósito práctico en el contexto municipal

En la vida real, clasificar predios por autovalúo sí tiene aplicaciones importantes:

✅ a) Gestión tributaria

El autovalúo determina cuánto debe pagar un propietario en Impuesto Predial.

Identificar qué predios son de “alto valor” ayuda a los municipios a:

Focalizar fiscalizaciones (por ejemplo, evitar que se subvalúen predios caros).

Planificar ingresos tributarios anuales.

Ejemplo: si sabes que el 25% de predios generan el 60% del total de autovalúo, puedes diseñar estrategias de cobranza más efectivas.

✅ b) Planificación urbana y social

La clasificación revela distribución socioeconómica de una ciudad.

Se pueden cruzar los predios “altos” con servicios públicos (agua, luz, seguridad) y ver desigualdades.

Ejemplo: en un distrito, las zonas con más predios de alto autovalúo tienen parques y pistas asfaltadas, mientras que las zonas con bajo autovalúo carecen de ellos → sirve para justificar proyectos de inversión.

```python
# ==== CREAR CLASE target_autovaluo ====
import pandas as pd
import numpy as np
import re, unicodedata

# 1) Cargar archivo limpio
ARCHIVO_LIMPIO = "predios_limpio_simple.csv"
datos = pd.read_csv(ARCHIVO_LIMPIO)

# 2) Funciones auxiliares para normalizar y convertir
def quitar_acentos(txt: str) -> str:
    return ''.join(c for c in unicodedata.normalize("NFD", str(txt)) if unicodedata.category(c) != "Mn")

def normalizar_nombre(txt: str) -> str:
    s = quitar_acentos(txt).lower()
    s = s.replace("\n", " ")
    s = re.sub(r"[^a-z0-9]+", "_", s)
    return re.sub(r"_+", "_", s).strip("_")

def convertir_a_numero(valor):
    if pd.isna(valor): return np.nan
    if isinstance(valor, (int, float)): return float(valor)
    s = str(valor).strip()
    s = s.replace("S/.", "").replace("S/", "").replace(",", "").replace("\xa0","")
    s = re.sub(r"\s+", "", s)
    try:
        return float(s)
    except:
        return np.nan

# 3) Detectar la columna de autovalúo
nombres_originales = list(datos.columns)
nombres_norm = [normalizar_nombre(c) for c in nombres_originales]

col_autov = None
for i, nn in enumerate(nombres_norm):
    if "autovaluo" in nn:
        col_autov = nombres_originales[i]
        break

if not col_autov:
    raise ValueError("No se encontró una columna que contenga 'autovaluo'.")

print("Columna de autovalúo detectada:", col_autov)

# 4) Convertir a numérico
datos[col_autov] = datos[col_autov].apply(convertir_a_numero)

# 5) Calcular Q3 y crear clase
umbral_q3 = datos[col_autov].quantile(0.75)
datos["target_autovaluo"] = (datos[col_autov] >= umbral_q3).astype(int)

# 6) Mostrar resumen
print("\n=== RESUMEN DE target_autovaluo ===")
print(f"Umbral Q3 de {col_autov}: {umbral_q3:,.2f}")
print("Distribución de la clase:")
print(datos["target_autovaluo"].value_counts().rename({0:"No alto (0)", 1:"Alto (1)"}))

# Ver primeras filas con autovalúo y clase
print("\nEjemplo de registros con autovalúo y target:")
print(datos[[col_autov, "target_autovaluo"]].head(10))

# 7) Guardar dataset final con clase
ARCHIVO_FINAL = "predios_con_target_autovaluo.csv"
datos.to_csv(ARCHIVO_FINAL, index=False, encoding="utf-8")
print(f"\n✔ Archivo con clase guardado en: {ARCHIVO_FINAL}")
```

```python
# ==== MODELO TIPO J48 (ÁRBOL ENTROPÍA) EXCLUYENDO EL AUTOVALÚO ====
import pandas as pd
import numpy as np
import re, unicodedata
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report

# 1) Cargar datos con la clase
ARCHIVO_FINAL = "predios_con_target_autovaluo.csv"
datos = pd.read_csv(ARCHIVO_FINAL)

# 2) Utilidades para detectar el nombre real de la columna de autovalúo (por si tiene tildes/saltos de línea)
def quitar_acentos(s):
    return ''.join(c for c in unicodedata.normalize("NFD", str(s)) if unicodedata.category(c) != "Mn")
def norm_col(s):
    s = quitar_acentos(s).lower().replace("\n"," ")
    s = re.sub(r"[^a-z0-9]+", "_", s)
    return re.sub(r"_+", "_", s).strip("_")

nombres_originales = list(datos.columns)
nombres_normal = [norm_col(c) for c in nombres_originales]

col_autovaluo = None
for i, nn in enumerate(nombres_normal):
    if "autovaluo" in nn:          # detecta columnas como "Autovalúo 2025 (S/)"
        col_autovaluo = nombres_originales[i]
        break
if not col_autovaluo:
    raise ValueError("No se encontró la columna de autovalúo. Revisa los nombres de columnas.")

# 3) Definir variables (X) y etiqueta (y)
y = datos["target_autovaluo"].astype(int)

# Excluir autovalúo, la clase y posibles IDs/códigos que no aportan patrón
posibles_ids = [c for c in datos.columns if any(tok in norm_col(c) for tok in ["id","codigo","cod","n_predio","numero_predio","n_"])]
excluir = set([col_autovaluo, "target_autovaluo"]) | set(posibles_ids)

X = datos[[c for c in datos.columns if c not in excluir]].copy()

# 4) Separar columnas numéricas y categóricas
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X.columns if c not in num_cols]

print("Columnas usadas para entrenar (sin autovalúo ni IDs):")
print(f"- Numéricas:   {len(num_cols)}")
print(f"- Categóricas: {len(cat_cols)}")

# 5) Partición entrenamiento/prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# 6) Preprocesamiento: One-Hot para categóricas + escalado para numéricas
prep_numericas = Pipeline(steps=[
    ("escala", StandardScaler(with_mean=False))
])
prep_categoricas = Pipeline(steps=[
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=True))
])

preprocesado = ColumnTransformer(
    transformers=[
        ("num", prep_numericas, num_cols),
        ("cat", prep_categoricas, cat_cols)
    ],
    remainder="drop"
)

# 7) Clasificador tipo J48 (entropía)
arbol = DecisionTreeClassifier(
    criterion="entropy",
    min_samples_leaf=5,
    random_state=42
)

modelo = Pipeline(steps=[
    ("prep", preprocesado),
    ("clf", arbol)
])

# 8) Entrenamiento
modelo.fit(X_train, y_train)

# 9) Predicción y métricas
y_pred = modelo.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
matriz = confusion_matrix(y_test, y_pred)

print("\n=== RESULTADOS DEL MODELO (Árbol tipo J48 ~ entropía) ===")
print(f"Accuracy: {accuracy:.4f}")
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred, digits=4, target_names=["No alto (0)","Alto (1)"]))

# 10) Matriz de confusión en texto
print("\nMatriz de Confusión (texto):")
print("               Predicho No alto (0)   Predicho Alto (1)")
print(f"Real No alto (0)       {matriz[0,0]:5d}              {matriz[0,1]:5d}")
print(f"Real Alto (1)          {matriz[1,0]:5d}              {matriz[1,1]:5d}")
```
